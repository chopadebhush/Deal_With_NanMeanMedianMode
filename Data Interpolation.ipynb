### Data Interpolation

    Data interpolation is process of esteemating data in range getting unknown values from known values


### Type of interpolation

1. Linear interpolation
2. cubic interpolation
3. polynomial interpolation

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.array([1,2,3,4,5])
y= np.array([1,3,5,7,9])
```

```python
plt.scatter(x,y)
```

```python
x_new =np.linspace(1,5,10)
y_inter =np.interp(x_new,x,y)
```

```python
x_new
```

```python
y_inter
```

```python
plt.scatter(x_new ,y_inter)
```

### Cubic Interpolation

```python
x =np.array([1,2,3,4,5])
y =np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
f =interp1d(x ,y,kind="cubic")

x_new =np.linspace(1,5 ,10)
y_int_cubix =f(x_new)
```

```python
plt.scatter(x_new ,y_int_cubix)
plt.show()
```

### Polynomial Interploation

```python
x =np.array([1,2,3,4,5])
y =np.array([1,4,9,1,25])
```

```python
polyfit =np.polyfit(x ,y ,2)
x_new =np.linspace(1,5,10)
y_interpolate =np.polyval(polyfit,x_new)
```

```python
plt.scatter(x_new,y_interpolate)
```
