#### Outlier treatment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary =[11,40,45,68,78,90,57,74,91,88,68,57,48,99,101,68,77,110,140,1000]
salary
```

```python
df =pd.DataFrame(salary,columns=["salary"])
```

```python
df
```

### What is capping

     Replace the outlier with nearest value but these values are not outlier.

```python
## implement capping concept
lower_cap =df.salary.quantile(0.05)
upper_cap =df.salary.quantile(0.95)
print("Lower cap is :",round(lower_cap,3))
print("Upper cap is :",round(upper_cap,3))
```

```python
df["Salary_capped"] =np.where(df.salary < lower_cap ,lower_cap ,np.where(df.salary > upper_cap,upper_cap ,df.salary))
```
